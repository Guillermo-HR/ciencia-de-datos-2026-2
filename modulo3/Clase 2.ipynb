Cambio de nombre de columnas

```python
ventas = ventas.rename(columns={"IdRegion": "CdRegion"})
```

```python
ventas = ventas.rename(columns={
    "PrecioUnitario": "CostoUnitario",
    "UnidadesVendidas": "ArtVendidos"
})
```

```python
ventas.columns = ['CdVendedor','FechaVenta','CdRegion','CdEstado','CdCiudad','Producto','CdFamilia','CdGenero','CdCategoria','CdTipoProducto','CdColor','PrecioUnitario','UnCdadesVendCdas','PorcentajeRentabilCdad','CdTipoVenta',]
```

```python
ventas.columns
```

Eliminaciones

```python
ventas.drop("CdVendedor", axis=1, inplace=True)
```

```python
ventas["VentaTotal"] = ventas["UnCdadesVendCdas"] * ventas["PrecioUnitario"]
```

```python

ventas["Ventas_Doble"] = ventas["VentaTotal"]
```

Agrupaciones

```python
productos_mas_vendido = ventas.groupby("Producto")["UnCdadesVendCdas"].mean()
productos_mas_vendido.head()
```

```python
productos_mas_vendido = ventas.groupby("Producto")["UnCdadesVendCdas"].sum()
productos_mas_vendido.head()
```

```python
print(ventas.groupby("Producto")["UnCdadesVendCdas"].agg(["sum","mean","max","min","count","size"]))
```
